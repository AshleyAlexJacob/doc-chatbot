```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
llm.invoke("Explain the theory of relativity in simple terms.")
```

```python
for chunk in llm.stream("Explain the theory of relativity in simple terms."):
    print(chunk.content, end='', flush=True)
```

```python

from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that translates {input_language} to {output_language}."),
    ("user", "Translate the following text: {text}"),
])
```

```python
chain = prompt | llm
```

```python
chain.invoke({
    "input_language": "English",
    "output_language": "Pashto",
    "text": "Hello, how are you?"
})
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful story teller that translates {input_language} to {output_language}. but creates a short story to the desired output language."),
    ("user", "Translate and genrate a short story for this topic: {text}"),
])
```

```python
chain = prompt | llm
chain.invoke({
    "input_language": "English",
    "output_language": "Pashto",
    "text": "Secret of Bermuda Triangle"
})
```
